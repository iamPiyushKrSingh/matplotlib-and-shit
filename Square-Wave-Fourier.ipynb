```python
import matplotlib.pyplot as plt
import numpy as np

L = 4
## x-values ##
x = np.linspace(-L, L, 10000)
## Terms in series ##
n_values = [1, 5, 10, 20, 50, 100]

def format_func(value, tick_number):
    # find number of multiples of pi/2
    N = int(np.round(4 * value / L))
    if N == 0:
        return "0"
    elif N == -1:
        return r"$-L/4$"
    elif N == 1:
        return r"$L/4$"
    elif N == 2:
        return r"$L/2$"
    elif N == -2:
        return r"$-L/2$"
    elif N % 4 > 0:
        return r"${0}L/4$".format(N)
    elif N==4:
        return r"$L$"
    elif N==-4:
        return r"$-L$"

## fig and axes defined ##
fig, axs = plt.subplots(3, 2, figsize=(25, 25), dpi = 300)

## main wave ##
def square_wave(x):
    if -L <= x < 0:
        return -1
    elif 0 <= x <= L:
        return 1

# y-array for main wave #
square_wave_y = [square_wave(x_i) for x_i in x]

## Fourier Solution ##
def fourier_series(x, n):
    series = 0
    for i in range(1, n+1, 2):
        series += 4/(i*np.pi)*np.sin(i*np.pi*x/L)
    return series

for i, ax in zip(n_values, axs.flatten()):
    ax.set_xlim(-L, L)
    ax.grid()
    ax.axhline(0, c='k')
    ax.axvline(0, c='k')
    ax.plot(x, square_wave_y, lw=3, c='r')
    ax.plot(x, fourier_series(x, i), label=f'$n = {i}$', c='b')
#     ax.xaxis.set_major_formatter(plt.FuncFormatter(format_func))
    
    ax.xaxis.set_major_locator(plt.MultipleLocator(L))
    ax.xaxis.set_major_formatter(plt.FuncFormatter(format_func))
    ax.set_xlabel(r"Position ($x$)", fontsize=28, labelpad=20)
    ax.set_ylabel(r"Displacement $y(x, 0)$", fontsize=28, labelpad=10)
    ax.tick_params(axis='x', labelsize=25)
    ax.tick_params(axis='y', labelsize=25)
#     ax.xaxis.set_minor_locator(plt.MultipleLocator(np.pi / 4))
    ax.legend(fontsize=25)
    
fig.tight_layout(pad=5)
plt.savefig("a-02_p-04-02.png")
plt.show()
```
